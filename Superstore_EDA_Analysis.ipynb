# Exploratory Data Analysis of Superstore Sales Dataset using Python

## Objective

The objective of this project is to perform Exploratory Data Analysis (EDA) on the Superstore dataset to identify sales trends, profitability patterns, and business insights using Python visualization and data analysis libraries.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries Imported Successfully!")
```

```python
df = pd.read_csv("superstore.csv", encoding='latin1')

df.head()
```

# Dataset Information

```python
df.info()
```

# Statistical Summary

```python
df.describe()
```

# Checking Duplicate Values

```python
df.duplicated().sum()
```

# Checking Missing Values

```python
df.isnull().sum()
```

# Sales by Category

```python
plt.figure(figsize=(10,6))

sns.barplot(x='Category', y='Sales', data=df)

plt.title('Sales by Category')

plt.show()

```

# Profit by Category

```python
plt.Figure(figsize=(8,5))

sns.barplot(x='Category', y='Profit', data=df)

plt.title('Profit by Category')

plt.show()
```

# Discount vs Profit

```python
plt.Figure(figsize=(8,5))

sns.scatterplot(x='Discount', y='Profit', data=df)

plt.title('Discount vs Profit')

plt.show()
```

# Monthly Sales Trend

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
monthly_sales = df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum()
```

```python
monthly_sales
```

```python
plt.figure(figsize=(8,5))

monthly_sales.plot()

plt.title('Monthly Sales Trend')

plt.xlabel('Month')
plt.ylabel('Total Sales')

plt.show
```

# Sales by Region

```python
plt.Figure(figsize=(8,5))

sns.barplot(x='Region', y='Sales', data=df)

plt.xlabel('Region')
plt.ylabel('Sales')

plt.title('Sales by Region')

plt.show()
```

# Sales by Sub-Category

```python
region_sales = df.groupby('Region')['Sales'].sum()

region_sales
```

```python
subcat_sales = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False)

subcat_sales
```

```python
plt.figure(figsize=(12,6))

subcat_sales.plot(kind='bar')

plt.title("Sales by Sub-Category")

plt.xlabel("Sub-Category")
plt.ylabel("Total Sales")

plt.show()
```

# Profit by Sub-Category

```python
subcat_profit = df.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False)

subcat_profit
```

```python
plt.figure(figsize=(12,6))

subcat_profit.plot(kind='bar')

plt.title("Profit by Sub-Category")

plt.xlabel("Sub-Category")
plt.ylabel("Total Profit")

plt.show()
```

# Final Conclusion

The Exploratory Data Analysis (EDA) conducted on the Superstore Sales dataset provided valuable insights into sales performance, profitability, customer demand, and regional trends.

The analysis revealed that the Technology category generated the highest sales and profit, making it the strongest-performing business segment. Sub-category analysis showed that Phones and Copiers were major contributors to revenue and profitability. In contrast, products such as Tables and Bookcases generated significant losses despite achieving comparatively high sales, indicating potential pricing or cost management issues.

The analysis also identified a strong negative relationship between discount levels and profit. Transactions with higher discounts frequently resulted in reduced or negative profit margins, suggesting that excessive discounting adversely impacts business profitability.

Time-series analysis demonstrated an overall increase in monthly sales from 2014 to 2017, reflecting positive business growth and strong market performance over time.

Overall, the project successfully transformed raw sales data into meaningful business insights using data cleaning, visualization, aggregation, and trend analysis techniques.

# Business Recommendations

1. Optimize Discount Strategy

High discount levels were associated with declining profitability. The company should consider limiting excessive discounts and implementing controlled promotional strategies to maintain healthy profit margins.

2. Focus on High-Performing Products

Sub-categories such as Phones, Copiers, and Accessories demonstrated strong profitability and sales performance. Increasing marketing efforts and inventory availability for these products may further improve revenue growth.

3. Investigate Loss-Making Products

Products like Tables and Bookcases generated negative profit despite high sales volume. The company should review pricing, operational costs, shipping expenses, and discount policies for these products.

4. Strengthen Regional Performance

Regional sales distribution appeared relatively balanced, which is a positive indicator of market stability. However, lower-performing regions can still be targeted with localized marketing strategies to improve growth opportunities.

5. Use Data-Driven Decision Making

Regular analysis of sales, profit, customer behavior, and discount impact can help the company make informed strategic decisions and improve long-term business performance.
